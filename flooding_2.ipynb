```python
import os
import folium
import json
from folium.features import CustomIcon
from folium.plugins import MousePosition
```

```python
flooding_range=os.path.join('data','flooding.png')
fld_line=os.path.join('data','fld_line.png')
fld_bar=os.path.join('data','fld_colorbar.png')

m=folium.Map([26.02,-80.1328],zoom_start=10.2,control_scale=True)
img1=folium.raster_layers.ImageOverlay(
    name='flooding range',
    image=flooding_range,
#     bounds=[[25.24958037,-80.4100463],[26.31958037,-80.07]],
    bounds=[[25.24902037,-80.4100463],[26.31902037,-80.07]],
    opacity=0.7,
    interactive=True,
    cross_origin=False,
    zindex=1,
)
img1.add_to(m)

img2=folium.raster_layers.ImageOverlay(
    name='flooding line',
    image=fld_line,
    bounds=[[25.24902037,-80.4100463],[26.31902037,-80.07]],
    opacity=1.0,
    interactive=True,
    cross_origin=False,
    zindex=1,
)
img2.add_to(m)

fld_icon=CustomIcon(
    fld_bar,
    icon_size=(80,250),
    icon_anchor=(40,125)
)
fld_marker=folium.Marker(
    location=[25.9,-80],
    icon=fld_icon
)
m.add_child(fld_marker)

folium.LayerControl().add_to(m)

kw = {
    'color': 'yellow',
    'line_cap': 'round',
    'fill': True,
   'fill_color': 'none',
    'weight': 2.5,
   'popup': 'Flooding Simulation area',
   'tooltip': '<strong>Click me!</strong>',
}

folium.Rectangle(
    bounds=[[25.2499537,-80.4100463], [26.3199537,-80.0241186]],
    line_join='round',
    dash_array='5, 5',
    **kw,
).add_to(m)

formatter = "function(num) {return L.Util.formatNum(num, 5) + '  ';};"

MousePosition(
    position='topright',
    separator=' ,  ',
    empty_string='NaN',
    lng_first=True,
    num_digits=20,
    prefix='Coordinates:',
    lat_formatter=formatter,
    lng_formatter=formatter
).add_to(m)

m.save(os.path.join('maps', 'FL_flooding.html'))
m
```
