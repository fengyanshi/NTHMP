```python
import os
import folium
import json
from folium.features import CustomIcon
```

```python
# flooding_range=os.path.join('data','foolding.png')
flooding_velocity=os.path.join('data','foolding_vel.png')

# fld_bar=os.path.join('data','fld_colorbar.png')
vel_bar=os.path.join('data','vel_colorbar.png')

m=folium.Map([25.8,-80.15],zoom_start=9,control_scale=True)
# img1=folium.raster_layers.ImageOverlay(
#     name='flooding range',
#     image=flooding_range,
#     bounds=[[25.2499537,-80.33],[26.3199537,-80.07]],
#     opacity=0.7,
#     interactive=True,
#     cross_origin=False,
#     zindex=1,
# )
# img1.add_to(m)

img2=folium.raster_layers.ImageOverlay(
    name='flooding velocity',
    image=flooding_velocity,
    bounds=[[25.2499537,-80.33],[26.3199537,-80.07]],
    opacity=0.7,
    interactive=True,
    cross_origin=False,
    zindex=1,
)
img2.add_to(m)

# fld_icon=CustomIcon(
#     fld_bar,
#     icon_size=(80,250),
#     icon_anchor=(40,125)
# )
# fld_marker=folium.Marker(
#     location=[25.9,-80],
#     icon=fld_icon
# )
# m.add_child(fld_marker)

vel_icon=CustomIcon(
    vel_bar,
    icon_size=(80,250),
    icon_anchor=(40,125)
)
vel_marker=folium.Marker(
    location=[25.9,-80],
    icon=vel_icon
)
m.add_child(vel_marker)

# folium.LayerControl().add_to(m)

m.save(os.path.join('maps', 'FL_flooding_velocity.html'))
m
```
